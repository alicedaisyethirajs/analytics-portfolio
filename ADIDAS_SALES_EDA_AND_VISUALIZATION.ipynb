```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("/content/Adidas US Sales Datasets.xlsx")
df.head()
```

```python
df.head()
```

```python
# we see that the dataset is a mess with column names messed up so lets clean it using the drop function
df=df[3:].reset_index(drop=True)   # skipping the first 3 rows
df.columns=df.iloc[0]              #  setting the column names to first row
df=df.drop(df.columns[0],axis=1)   # dropping the empty columns
df.head()
```

```python
# lets convert he column names to lower case and standard form
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.head()
```

```python
print(" dataset info:")
print(df.info())
```

1.   convert price per unit,units sold,total sales,operating profit,operating margin to numeric
2.   it is seen that the invoice date is in object dtype so lets convert it into datetime dtype and add weekday , day, month , year and if its a weekend for better and concise analysis and report



```python
df["price_per_unit"]=pd.to_numeric(df["price_per_unit"],errors='coerce')
df["units_sold"]=pd.to_numeric(df["units_sold"],errors='coerce')
df["total_sales"]=pd.to_numeric(df["total_sales"],errors='coerce')
df["operating_profit"]=pd.to_numeric(df["operating_profit"],errors='coerce')
df["operating_margin"]=pd.to_numeric(df["operating_margin"],errors='coerce')
```

```python
df.info()
```

```python
# now on the dates and adding new columns for the dates
df=df.drop(index=0)
df["invoice_date"]=pd.to_datetime(df["invoice_date"],errors='coerce')
df["weekday"]=df["invoice_date"].dt.day_name()
df["day"]=df["invoice_date"].dt.day
df["month"]=df["invoice_date"].dt.month
df["year"]=df["invoice_date"].dt.year

def weekend(x):
  return "yes" if x in ["Satueday","Sunday"] else "no"
df["is_weekend"]=df["weekday"].apply(weekend)


df.head()
```

```python
print("statistics:")
print(df.describe())
```

```python
print("missing values:")
print(df.isnull().sum())
```

```python
#select only column numerical column names
numerical_columns=df.select_dtypes(include=[np.number]).columns
len(numerical_columns)

num_vars=len(numerical_columns)

#fixing no of columns in subplot to 2
cols=2
rows= (num_vars// cols)+(num_vars%cols>0)

fig,axes=plt.subplots(rows,cols, figsize=(15,5*rows))

axes=axes.flatten()

# loop for visualization
for i, col in enumerate(numerical_columns):
  sns.histplot(df[col],ax=axes[i],kde=True)
  axes[i].set_title(f'Distribution of {col}')
  axes[i].set_ylabel('Frequency')

plt.tight_layout(pad=5)
plt.show()
```

```python
#columns we want to plot
column=['retailer','region','state','city','product']
len(column)
cols=2
rows=(num_vars//cols)+(num_vars%cols>0)

fig,axes=plt.subplots(rows,cols,figsize=(15,5*rows))
axes=axes.flatten() # flattens the grid to 1d for easy looping


for i, col in enumerate(column):
  sns.countplot(df[col],ax=axes[i])
  axes[i].set_title(f'Distribution of {col}')
  axes[i].set_ylabel('Frequency')
  axes[i].set_xticklabels(axes[i].get_xticklabels(),rotation=45,ha='right')
#deletes an empty subplot creates
for j in range(i+1,len(axes)):
  fig.delaxes(axes[j])

plt.tight_layout(pad=3)
plt.show()
```
